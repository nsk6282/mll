```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df_setosa = df.loc[df['species']=='setosa']
```

```python
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
df_virginica = df.loc[df['species']=='virginica']
```

```python
df_versicolor.head()
```

```python
len(df)
```

```python
plt.plot(df_setosa['petal_length'],np.zeros(len(df_setosa)),'o')
plt.plot(df_versicolor['petal_length'],np.zeros(len(df_versicolor)),'o')
plt.plot(df_virginica['petal_length'],np.zeros(len(df_virginica)),'o')
plt.xlabel("Petal Length")
plt.show()

```

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

```python
sns.jointplot("petal_length","sepal_width",df,kind='hex')
```

```python
sns.pairplot(df,hue='species',palette='coolwarm')
```

```python
dfc=df.corr()
sns.heatmap(dfc, annot= True,)
```
